```python
def regresion(x):
    y= (2.9**(-7))*x**3 - (1.3**(-4))*x**2 - (1.4**(-4))*x + 5891.25
    return y
```

```python
L1=zeros(len(p1),dtype="float")
L2=zeros(len(p1),dtype="float")
L3=zeros(len(p1),dtype="float")
L4=zeros(len(p1),dtype="float")
L5=zeros(len(p1),dtype="float")
```

```python
for i in range(len(p1)):
    L1[i]=regresion(p1[i])
    L2[i]=regresion(p2[i])
    L3[i]=regresion(p3[i])
    L4[i]=regresion(p4[i])
    L5[i]=regresion(p5[i])
```

```python
plot(c1,L1)
xlabel("$ longitud\ de\ onda\ \lambda $",size=15)
ylabel("$ count $",size=15)
title("$ espectro\ 1\ $",size=15)

```

```python
plot(c1,L1,"-o",c="c",label="$ espectro\ 1\ $")
plot(c5,L5,"-o",c="b",label="$ espectro\ 2\ $")
xlabel("$ longitud\ de\ onda\ \lambda $",size=15)
ylabel("$ count $",size=15)
title("$ espectro\ 1\ y\ 5\ $",size=15)
xlim(300,400)
legend()
```

```python
def minimos(L1,j):
    min1=zeros(40)
    min2=zeros(60)
    for i in range(300,340):
        min1[i-300]=L1[i]
    
    for i in range(340,400):
        min2[i-340]=L1[i]
    print "espectro numero", j
    print "primer minimo",min(min1),"segundo minimo",min(min2),min(min1)-min(min2),"\n"
    return min(min1),min(min2)
```
