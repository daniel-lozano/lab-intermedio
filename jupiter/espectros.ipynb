```python

def regresion(x):
    y= -(3.6*10**(-9))*x**3 - (3.2*10**(-5))*x**2 + (0.277)*x + 5801.092
    return y
```

```python
def ponderar(mini,p,C):
    pon=2
    m=0
    s=0
    for i in range(pon):
        s=s+C[mini-pon/2+1+i]
        m=m+C[mini-pon/2+1+i]*p[mini-pon/2+i]
    print m/s    
    return m/s
```
