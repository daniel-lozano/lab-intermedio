```python

def regresion(x):
    y= -(6.6317*10**(-9))*x**3 - (3.132*10**(-5))*x**2 + (0.277)*x + 5801.0832
    return y
```

```python
L1=zeros(len(p1),dtype="float")
L2=zeros(len(p1),dtype="float")
L3=zeros(len(p1),dtype="float")
L4=zeros(len(p1),dtype="float")
L5=zeros(len(p1),dtype="float")
```

```python
for i in range(len(p1)):
    L1[i]=regresion(p1[i])
    L2[i]=regresion(p2[i])
    L3[i]=regresion(p3[i])
    L4[i]=regresion(p4[i])
    L5[i]=regresion(p5[i])
```
