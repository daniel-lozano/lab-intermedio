```python
w=loadtxt(open("datos_thor.txt","r"))
```

```python
pixels=w[:,0]
count=w[:,1]
```

```python
p=[]
c=[]
for i in range(len(count)):
    if(count[i]>5000):
        append(c,count[i])
        append(p,pixels[i])
```

```python
plot(pixels,count)

xlabel("$ pixels\ $",size=20)
ylabel("$ count\ $",size=20,rotation="horizontal")
title("$ thorium\ espectro\ de\ calibracion\  $",size=22)
#figsize()
scatter(pixels[479],count[479],c="r",s=30)
scatter(pixels[531],count[531],c="r",s=30)
scatter(pixels[556],count[556],c="r",s=30)
scatter(pixels[682],count[682],c="r",s=30)
scatter(pixels[688],count[688],c="r",s=30)
scatter(pixels[751],count[751],c="r",s=30)
scatter(pixels[759],count[759],c="r",s=30)
scatter(pixels[774],count[774],c="r",s=30)
xlim(min(pixels),max(pixels))
ylim(min(count),max(count)+1000)
savefig("espec_thor.png")
```
