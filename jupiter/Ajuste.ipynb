# regresion polinomica polifit

```python
a,b,c,d=polyfit(pix,lambd,3)
```

```python
x=linspace(0,900)
y=zeros(len(x),dtype="float")
```

```python
for i in range(len(x)):
    y[i]=a*x[i]**3+b*x[i]**2+c*x[i]+d
```
